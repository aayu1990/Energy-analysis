```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
coalData = pd.read_csv("Data_new_2\\Coal_Production,_Annual.csv")
crudeData = pd.read_csv("Data_new_2\\Crude_Oil_Production,_Annual.csv")
dryNatGasData = pd.read_csv("Data_new_2\\Natural_Gas_(Dry)_Production,_Annual.csv")
wetNatGasData = pd.read_csv("Data_new_2\\Natural_Gas_Plant_Liquids_Production,_Annual.csv")
biomassData = pd.read_csv("Data_new_2\\Total_Biomass_Energy_Production,_Annual.csv")
fossilData = pd.read_csv("Data_new_2\\Total_Fossil_Fuels_Production,_Annual.csv")
windData = pd.read_csv("Data_new_2\\Wind_Energy_Consumption-Production,_Annual.csv")
solarData = pd.read_csv("Data_new_2\\Solar-PV_Energy_Consumption-Production,_Annual.csv")
nuclearData = pd.read_csv("Data_new_2\\Nuclear_Electric_Power_Consumption-Production,_Annual.csv")
hydroData = pd.read_csv("Data_new_2\\Hydroelectric_Power_Consumption-Production,_Annual.csv")
geoData = pd.read_csv("Data_new_2\\Geothermal_Energy_Consumption-Production,_Annual.csv")
```

```python
for x in range(1, 5):
    dataDic = {'Energy Type' :
    ['Coal',
    'Crude Oil',
    'Dry Natural Gas',
    'Liquid Natural Gas',
    'Biomass',
    'Fossil Fuel',
    'Wind',
    'Solar',
    'Nuclear',
    'Hydroelectric',
    'Geothermal'],

    'Energy in Trillions of Btu' :
    [float(round(coalData.loc[coalData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(crudeData.loc[crudeData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(dryNatGasData.loc[dryNatGasData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(wetNatGasData.loc[wetNatGasData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(biomassData.loc[biomassData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(fossilData.loc[fossilData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(windData.loc[windData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(solarData.loc[solarData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(nuclearData.loc[nuclearData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(hydroData.loc[hydroData["Year"] == (2010 + x)]['Trillion Btu'],2)),
    float(round(geoData.loc[geoData["Year"] == (2010 + x)]['Trillion Btu'],2))]}
    
    
    dataDicPd = pd.DataFrame.from_dict(dataDic)
    dataDicPd.plot(kind='pie',colors=('#1CFF0D','#00E84E', '#00FFB9', '#0CD1E8', '#0D8EFF', '#25FF3C', '#2DE88A', '#31FFE7', '#21B6E8', '#2579FF', '#6895FF'), y = 'Energy in Trillions of Btu', autopct='%1.1f%%', 
    startangle=0, shadow=False, labels=dataDicPd['Energy Type'], fontsize=14)
    plt.ylabel("")
    plt.legend(loc='right', bbox_to_anchor=(1.7, 0.5))
    
    plt.savefig("Images/Primary_Energy_Consumption_Types" + str((2010 + x)) + ".png")
```
