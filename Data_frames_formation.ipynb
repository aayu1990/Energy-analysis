```python
import pandas as pd
import os
import csv
from io import StringIO
import matplotlib.pyplot as plt
from os import path
```

```python
dfs=[]
counter=0
for file in os.listdir('Data_new_2'):
    if file.endswith(".csv"): 
        open_file=open('Data_new_2/' + file,'r') 
        content = open_file.read()
        f = StringIO(content)
        reader = csv.reader(f, delimiter=',')
        arr = []
        for row in reader:
            arr.append(row)
        df=pd.DataFrame(arr)
        counter+=1
        print(f"data frame number: {counter}")
        print(file)
        print(df)
        dfs.append(df)
        continue
    else:
        continue
print(f"\ntotal {len(dfs)} data frames")

```

```python
#take the above data frames and plot, good luck
fig = plt.figure()
outpath = "../Plots"
for data_frame in dfs:
    x_vals=data_frame[0][1:]
    y_vals=data_frame[1][1:]
    plt.subplots(figsize=(40, 30))
    plt.scatter(x_vals, y_vals,marker="o",edgecolors="black",alpha=1)
    axis_font = {'fontname':'Arial', 'size':'18'}
#     plt.title(f"{file}")
    ylabel = file
    ylabel=file.replace('.csv', ' - ')
    ylabel=ylabel+ data_frame[1][0]
    plt.ylabel(ylabel,**axis_font)
    plt.grid(True)
    
  
    
```
