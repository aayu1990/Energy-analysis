```python
import pandas as pd
import os
import csv
from io import StringIO
import matplotlib.pyplot as plt
from os import path
```

```python
dfs=[]
counter=0
for file in os.listdir('Data_new_2'):
    if file.endswith(".csv"): 
        open_file=open('Data_new_2/' + file,'r') 
        content = open_file.read()
        f = StringIO(content)
        reader = csv.reader(f, delimiter=',')
        arr = []
        for row in reader:
            arr.append(row)
        df=pd.DataFrame(arr)
        counter+=1
        print(f"data frame number: {counter}")
        print(file)
        print(df)
        dfs.append(df)
        continue
    else:
        continue


```
